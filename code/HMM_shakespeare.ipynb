```python
import nltk
import numpy as np
import sklearn
import copy
```

```python
lines = []
with open('../data/shakespeare.txt') as inp:
      for line in inp:
        if len(line.split()) > 2:
            lines.append(line.lower())
            
```

```python
for i in range(len(lines)):
    lines[i] = lines[i][:-1]
    lines[i] = lines[i].replace("'s ", " ")
    lines[i] = lines[i].replace("(", "")
    lines[i] = lines[i].replace(")", "")
    lines[i] = lines[i].replace(":", "")
    lines[i] = lines[i].replace(",", "")
    lines[i] = lines[i].replace("'", "")
    lines[i] = lines[i].replace("?", "")
    lines[i] = lines[i].replace(".", "")
    lines[i] = lines[i].replace("!", "")
    lines[i] = lines[i].replace(";", "")
```

```python
print(lines)
```

```python
t_lines = []
for i in range(len(lines)):
    t_lines.append(nltk.tokenize.word_tokenize(lines[i]))
```

```python
print(t_lines)
```

```python
word_d = {}
counter = 0
for line in t_lines:
    for word in line:
        if word not in word_d:
            word_d[word] = counter
            counter += 1
```

```python
wd2 = {}
for key in word_d:
    wd2[word_d[key]] = key

def nums_to_words(line):
    ret = []
    for num in line:
        if num == 10000:
            ret.append('fuck')
        else:
            ret.append(wd2[num])
        
    return ret
```
