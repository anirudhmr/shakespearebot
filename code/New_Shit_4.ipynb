```python
lines = []
with open('../data/shakespeare.txt') as inp:
      for line in inp:
        if len(line.split()) > 2:
            lines.append(line)
            
```

```python
for i in range(len(lines)):
    lines[i] = lines[i][:-1]
    #lines[i] = lines[i].replace("'", "AnkushHD")
    #lines[i] = lines[i][:-1]
    lines[i] = lines[i].replace("'s ", " ")
    lines[i] = lines[i].replace("(", "")
    lines[i] = lines[i].replace(")", "")
    lines[i] = lines[i].replace(":", "")
    lines[i] = lines[i].replace(",", "")
```

```python
print(lines)
```

```python
t_lines = []
for i in range(len(lines)):
    t_lines.append(nltk.tokenize.word_tokenize(lines[i]))
    #t_lines.append(nltk.tokenize.TweetTokenizer(lines[i]))
```

```python
print(t_lines)
```

```python
word_d = {}
counter = 0
for line in t_lines:
    for word in line:
        if word not in word_d:
            word_d[word] = counter
            counter += 1
```

```python
t_lines_2 = copy.deepcopy(t_lines)
for i in range(len(t_lines_2)):
    for j in range(len(t_lines_2[i])):
        t_lines_2[i][j] = word_d[t_lines_2[i][j]]
```
